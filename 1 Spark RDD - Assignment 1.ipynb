```
#map - Return a new distributed dataset formed by passing each element of the source through a function func.
print(rdd.collect())
#sum of 2 numbers in rdd.
print(rdd.map(lambda x: x[0]+x[1]).collect())
```

```
#filter - 	Return a new dataset formed by selecting those elements of the source on which func returns true.

# only elements if second element is 2 times. 
def filt(x):
  if x[0] == x[1]//2:
    return True
  return False

rdd.filter(filt).collect()
```

```
# flatMap - 	Similar to map, but each input item can be mapped to 0 or more output items (so func should return a Seq rather than a single item).
#rdd_Flat = sc.parallelize([[1,2,3,2,4,5
print(rdd.collect())
rdd.flatMap(lambda x: [x[0] + x[1]]).collect()
```

```
# mapPartitions - Similar to map, but runs separately on each partition (block) of the RDD, so func must be of type Iterator<T> => Iterator<U> when running on an RDD of type T.

def func(partition):
  for element in partition:
    print(element)
    #sum of element in tuple divisible by 2
    if (element[0]+element[1])%2 == 0:
      yield True
    else:
      yield False

      
rdd = sc.parallelize([(1,2),(2,4),(4,6)])
#print(rdd.collect())
rdd.mapPartitions(func).collect()
```

```
# mapPartitionsWithIndex - Similar to mapPartitions, but also provides func with an integer value representing the index of the partition, so func must be of type (Int, Iterator<T>) => Iterator<U> when running on an RDD of type T.

def mpwi(splitIndex, partition):
  yield splitIndex

rdd = sc.parallelize([(1,2),(2,4),(4,6)])
rdd.mapPartitionsWithIndex(mpwi).collect()

```

```
# aggregateByKey - When called on a dataset of (K, V) pairs, returns a dataset of (K, U) pairs where the values for each key are aggregated using the given combine functions and a neutral "zero" value.

premierRDD = sc.parallelize([
  ("Arsenal", "2014–2015", 75), ("Arsenal", "2015–2016", 71), ("Arsenal", "2016–2017", 75), ("Arsenal", "2017–2018", 63),
 ("Chelsea", "2014–2015", 87), ("Chelsea", "2015–2016", 50), ("Chelsea", "2016–2017", 93), ("Chelsea", "2017–2018", 70), 
 ("Liverpool", "2014–2015", 62), ("Liverpool", "2015–2016", 60), ("Liverpool", "2016–2017", 76), ("Liverpool", "2017–2018", 75),
 ("M. City", "2014–2015", 79), ("M. City", "2015–2016", 66), ("M. City", "2016–2017", 78), ("M. City", "2017–2018", 100), 
 ("M. United", "2014–2015", 70), ("M. United", "2015–2016", 66), ("M. United", "2016–2017", 69), ("M. United", "2017–2018", 81) 
 ])

def seqFunc(acc, teams):
 if(acc[1] > teams[1]):
   return acc 
 else: 
   return teams
 
def combFunc(acc1, acc2):
 if(acc1[1] > acc2[1]):
   return acc1 
 else:
   return acc2
  
  
premierMap = premierRDD.map(lambda t: (t[0], (t[1], t[2])))

print(premierMap.collect())


premierAgg = premierMap.aggregateByKey((0,0), seqFunc, combFunc)

premierAgg.collect()


```

```
# join - When called on datasets of type (K, V) and (K, W), returns a dataset of (K, (V, W)) pairs with all pairs of elements for each key. Outer joins are supported through leftOuterJoin, rightOuterJoin, and fullOuterJoin.
rdd_a = sc.parallelize([('a',1), ('b',2)])
rdd_b = sc.parallelize([('a',3), ('a',5)])
rdd_a.join(rdd_b).collect()
```

```
# cogroup - When called on datasets of type (K, V) and (K, W), returns a dataset of (K, (Iterable<V>, Iterable<W>)) tuples. This operation is also called groupWith.
print([(x, tuple(map(list, y))) for x, y in sorted(list(rdd_a.cogroup(rdd_b).collect()))])
```

```
# pipe - Pipe each partition of the RDD through a shell command, e.g. a Perl or bash script. RDD elements are written to the process's stdin and lines output to its stdout are returned as an RDD of strings.
rdd.pipe('cat').collect()
```

```
# coalesce - Decrease the number of partitions in the RDD to numPartitions. Useful for running operations more efficiently after filtering down a large dataset.
rdd_gbk = sc.parallelize([(2,1), (1,2), (1,3), (5,8)])
print(rdd_gbk.getNumPartitions())
rdd_gbk = rdd_gbk.coalesce(2)
print(rdd_gbk.getNumPartitions())
```

```
# repartition - Reshuffle the data in the RDD randomly to create either more or fewer partitions and balance it across them. This always shuffles all data over the network.
rdd_gbk = sc.parallelize([('a',1), ('b',2), ('a',3), ('c',8)])
print(rdd_gbk.getNumPartitions())
rdd_gbk = rdd_gbk.repartition(4)
print(rdd_gbk.getNumPartitions())
#print(rdd_a.getNumPartitions())
```

```
# repartitionAndSortWithinPartitions - 	Repartition the RDD according to the given partitioner and, within each resulting partition, sort records by their keys. This is more efficient than calling repartition and then sorting within each partition because it can push the sorting down into the shuffle machinery.

rdd_gbk = sc.parallelize([(2,1), (1,2), (1,3), (5,8)])
print(rdd_gbk.getNumPartitions())
rdd_gbk = rdd_gbk.repartitionAndSortWithinPartitions(4)
print(rdd_gbk.getNumPartitions())
print(rdd_gbk.collect())
```
